```python
# Dependencies
import pandas as pd
import numpy as np
from numpy import NaN
import matplotlib.pyplot as plt
import requests
import json
import gmaps
from config import gkey
```

```python
part1 = "Resources/part1.csv"
part2 = "Resources/part2.csv"
part3 = "Resources/part3.csv"
part4 = "Resources/part4.csv"
part5 = "Resources/part5.csv"
part1_df = pd.read_csv(part1)
part2_df = pd.read_csv(part2)
part3_df = pd.read_csv(part3)
part4_df = pd.read_csv(part4)
part5_df = pd.read_csv(part5)
```

```python
# Concatinate all of the DataFrames into one

Frames = [part1_df,part2_df,part3_df,part4_df,part5_df]
Mexico_Gas_Stations = pd.concat(Frames)
Mexico_Gas_Stations
```

# MAPAS
* TOP 100 best rating
* TOP 100 worst rating 
* TOP 100 minium prices

```python
#Get TOP 1000 of best valuated stations in descendent order by Rating and then by User Raitings 
best_rating_df = Mexico_Gas_Stations3.sort_values(by="Rating",ascending=False)
best_rating_df = best_rating_df.sort_values(by="User Ratings",ascending=False).head(100)
best_rating_df
```

```python
#Set api key
gmaps.configure(api_key="g_key")
#Set lat and lng as locations
locations = best_rating_df[["Lat","Lng"]]
#Set gas_price as weight
weight=best_rating_df["Rating"]
```

```python
# Plot Best Rating Heatmap
fig = gmaps.figure()
max_intensity = np.max(weight)

# Create heat layer
heatmap = gmaps.heatmap_layer(
    locations, 
    weights = weight, 
    dissipating=False, 
    point_radius=.2)

# Add layer
fig.add_layer(heatmap)

# Display figure
fig
```

```python
#Get TOP 1000 of worst valuated stations in descendent order by Rating and then by User Raitings 
worst_rating_df = Mexico_Gas_Stations3.sort_values(by="Rating",ascending=True)
# worst_rating_df = worst_rating_df.sort_values(by="User Ratings",ascending=False).head(100)
worst_rating_df
```

```python
Mexico_Gas_Stations3 = Mexico_Gas_Stations3.loc[Mexico_Gas_Stations3["User Ratings"]>0]
Mexico_Gas_Stations3 = Mexico_Gas_Stations3.loc[Mexico_Gas_Stations3["gas_price"]>16.45]
min_precio_prueba_mapaCalor_df = Mexico_Gas_Stations3.sort_values(by="Rating",ascending=False)
min_precio_prueba_mapaCalor_df = min_precio_prueba_mapaCalor_df.sort_values(by="User Ratings",ascending=False)
min_precio_prueba_mapaCalor_df = min_precio_prueba_mapaCalor_df.sort_values(by="gas_price",ascending=True).head(100)
# min_precio_prueba_mapaCalor_df = min_precio_prueba_mapaCalor_df.sort_values(by="Rating",ascending=False)
# min_precio_prueba_mapaCalor_df = min_precio_prueba_mapaCalor_df.sort_values(by="User Ratings",ascending=False).head(100)
min_precio_prueba_mapaCalor_df
```

```python
#Set api key
gmaps.configure(api_key=g_key)
#Set lat and lng as locations
locations = min_precio_prueba_mapaCalor_df[["Lat","Lng"]]
#Set Humidity as weight
weight=min_precio_prueba_mapaCalor_df["gas_price"]
# #Plot Heatmap
# fig = gmaps.figure
# max_intensity = np.max(weight)
```

```python
# Plot Heatmap
fig = gmaps.figure()
max_intensity = np.max(weight)

# Create heat layer
heatmap = gmaps.heatmap_layer(
    locations, 
    weights = weight, 
    dissipating=False, 
    max_intensity=17, 
    point_radius=.2)

# Add layer
fig.add_layer(heatmap)

# Display figure
fig
```

```python
# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{name}</dd>
<dt>City</dt><dd>{Address}</dd>
<dt>Rating</dt><dd>{Rating}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
gas_info = [info_box_template.format(**row) for index, row in min_precio_prueba_mapaCalor_df.iterrows()]
locations = min_precio_prueba_mapaCalor_df[["Lat", "Lng"]]
# Add marker layer ontop of heat map
markers = gmaps.marker_layer(locations, info_box_content = gas_info)
markers
# Display Map
fig.add_layer(markers)

fig
```
